```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow

try:
    img = cv2.imread('/content/OIP.webp')
    if img is None:
        raise FileNotFoundError("Image file not found.")
except FileNotFoundError as e:
    print(f"Error: {e}")
    print("Please make sure 'input_image.jpg' exists in the same directory or provide the full path.")
    img = np.zeros((200, 300, 3), dtype=np.uint8)
    cv2.putText(img, "Image not found", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)


gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(gray_img.ravel(), 256, [0, 256])
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

equalized_img = cv2.equalizeHist(gray_img)

plt.subplot(1, 2, 2)
plt.hist(equalized_img.ravel(), 256, [0, 256])
plt.title('Histogram of Equalized Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()


clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_img = clahe.apply(gray_img)

cv2_imshow(img)
cv2_imshow(equalized_img)
cv2_imshow(clahe_img)
```
