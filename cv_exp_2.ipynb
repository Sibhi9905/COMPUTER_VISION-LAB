```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
for name, data in uploaded.items():
    image = cv2.imdecode(np.frombuffer(data, np.uint8), cv2.IMREAD_COLOR)
    break

alpha = 1.8
beta = 128 - 128 * alpha
adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
if image.shape != adjusted.shape:
    adjusted = cv2.resize(adjusted, (image.shape[1], image.shape[0]))

combined = np.hstack((image, adjusted))
print("Original Image (Left) vs Contrast Adjusted Image (Right):")
cv2_imshow(combined)
cv2.imwrite('combined_result.jpg', combined)
```
